Заполнение пропусков

```python
data['bill_length_mm'] = data['bill_length_mm'].fillna(data['bill_length_mm'].median())
data['bill_depth_mm'] = data['bill_depth_mm'].fillna(data['bill_depth_mm'].median())
data['flipper_length_mm'] = data['flipper_length_mm'].fillna(data['flipper_length_mm'].median())
data['body_mass_g'] = data['body_mass_g'].fillna(data['body_mass_g'].median())
```

```python
data['sex'] = data['sex'].fillna('Unknown')
```

```python
data.isnull().sum()
```

Нормализация

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

data[['mm_bill_length','mm_body_mass']] = scaler.fit_transform(
    data[['bill_length_mm','body_mass_g']]
)
```

Проверка нормализации

```python
fig, axs = plt.subplots(nrows=2)

sns.lineplot(data=data, x='bill_length_mm', y='body_mass_g', ax=axs[0])
sns.lineplot(data=data, x='mm_bill_length', y='mm_body_mass', ax=axs[1])
```

Кодирование категориальных признаков

```python
from sklearn.preprocessing import OrdinalEncoder

enc = OrdinalEncoder()

data[['sex']] = enc.fit_transform(data[['sex']])
```

```python
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder()

res = enc.fit_transform(data[['island']])

res.toarray()
```

```python
data[enc.categories_[0]] = res.toarray()
```

PCA — уменьшение размерности

```python
from sklearn.decomposition import PCA
```

```python
features = data[['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g']]
```

```python
pca = PCA(n_components=2)
components = pca.fit_transform(features)
```

```python
pca_df = pd.DataFrame(components, columns=['PC1','PC2'])
```

```python
sns.scatterplot(data=pca_df, x='PC1', y='PC2')
```
